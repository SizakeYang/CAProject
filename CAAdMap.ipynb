```python
import folium
import requests
import numpy as np
import pandas as pd
from folium.plugins import HeatMap
```

```python
def getLocation(text):
    url = "http://api.map.baidu.com/geocoding/v3/"
    payload = {
        'output':'json',
        'ak':'wQ6MXrttpBaDqGxhBDKN4wkaIijnAzEL',
        'city':'汕头市'
        }
    payload['address'] = text
    content = requests.get(url,params=payload,timeout=(5, 5)).json()
    print(content['result']['location'])
    return content['result']['location']
```

```python
getLocation('汕头市金平区')
# {'lng': 116.70955006744148, 'lat': 23.371750385773453}
```

```python
data['location'] = data['pre_address'].apply(getLocation)
data['location']
```

```python
lon = np.array([i["lng"] for i in data['location']],dtype=float)
lat = np.array([i["lat"] for i in data['location']],dtype=float)
#scale = np.array(data['punishAmount'])
scale = np.ones(len(lat),float)
```

```python
data_m = [[lat[i],lon[i],scale[i]] for i in range(len(scale))]
```
